## Task
What are the different modes of transport which are operated by Transfer for London? How many of modes do they have?

Print the list with different modes of transport, plus their count. Example output:
```
[bus, cable-car,.....]
Number of different modes of transport is: xyz
```

We need to search the documentation for correct request.

````python
help(requests.get)
````

````python
url = "https://api.tfl.gov.uk/Line/Route"
query = "serviceTypes"
````

````python
res1 = requests.get(url + "?" + query)
````

````python
json_res1 = res1.json()
````

````python
modes_of_transport = []

for route in json_res1:
    if route["modeName"] not in modes_of_transport:
        modes_of_transport.append(route["modeName"])

print(modes_of_transport)
print(f"number of different modes of transport is {len(modes_of_transport)}")
````

## Task
How many BikePoints in London are operated by Transfor for London? How many docks are in **all** BikePoints? There is the information for empty and full docks for each BikePoint.

````python
url = "https://api.tfl.gov.uk/BikePoint/"
````

````python
res2 = requests.get(url)
````

````python
json_res2 = res2.json()
````

````python
nbBikepoints = 0
docks_total = 0

for bikepoint in json_res2:
    nbBikepoints += 1
    nbDocks = bikepoint["additionalProperties"][8]["value"]
    nbDocks = int(nbDocks)
    docks_total += nbDocks

print(nbBikepoints)
print(docks_total)
````

## Task
How many tube and bus lines are in London? Print names of all tube lines.

````python
url = "https://api.tfl.gov.uk/Line/Mode/bus%2Ctube"
````

````python
res3 = requests.get(url)
````

````python
lines = res3.json()
````

````python
tube_line_names = []
nbTubeLines = 0
nbBusLines = 0

for line_info in lines:
    if line_info["modeName"] == "bus":
        nbBusLines += 1
    elif line_info["modeName"] == "tube":
        nbTubeLines += 1
        tube_line_names.append(line_info["name"])
    else:
        print("modeName Error")


print(nbTubeLines)
print(nbBusLines)
for name in tube_line_names: print(name)
````
